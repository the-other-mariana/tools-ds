# Programming in Matplotlib

## Layers

1. Scripting (pyplot): quick plots with no style

2. Artistic: ticks, axis, text

3. Backend: involves details about the rendering of plots, for custom plots

    1. dynamic plots
    2. interactive
    3. js/html
    
## Artistic Layer

1. scripting: plt.plot()

2. artistic: plt.subplot(), which returns two objects:
    
    1. figure
    2. axes
    
3. artistic: plt.subplots(), which returns two objects:

    1. figure
    2. list of axes

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# show default mpl params and everything we can modify
mpl.rcParams
```

```python
plt.plot((1,2))
```

```python
# modify mpl params for all plots
mpl.rcParams["figure.figsize"] = (8, 4)
plt.plot((3,1,2,4)) # if you pass only a list, it will plot heights
```

```python
# returns to default values
mpl.rcdefaults()
```

## Modify the parameters for just one plot in particular

```python
x = [1,2,3,4,5]
y = [3,2,1,4,5]
# returns the object that controls the data of the figure
fig = plt.figure() # 0 axes # creates an empty space of figsize
ax1 = fig.add_subplot(1,1,1) # returns axes # location: total rows, total cols, index of this plot
ax1.plot(x,y)
# before, you drew on the default plot of mpl, now you create yours and have objects you can modify
```

```python
fig = plt.figure()
fig.subplots(2,2)
```

```python
# option 1
fig = plt.figure()
(ax1, ax2), (ax3, ax4) = fig.subplots(2,2) # as it returns two lists
ax1.plot([1,2,3], color='red')
ax2.plot([3,2,1], color='blue')
ax3.plot([4,4,4], color='orange')
ax4.plot([1,0,1], color='black')
fig.tight_layout() # you can put plt.tight_layout() to modify mpl default canvas, or yours with fig
```

```python
# option 2
rows = 2
cols = 3
fig, axes = plt.subplots(rows, cols)
axes[0][0].plot(np.arange(10), 'x')
```

```python
rows = 2
cols = 3
fig, axes = plt.subplots(rows, cols)
for i in range(rows):
    for j in range(cols):
        axes[i][j].plot(np.arange(i + j + 1), 'x')
```

```python
# option 3
fig = plt.figure( figsize=(10,5) )
grid = fig.add_gridspec(3,3)
ax1 = fig.add_subplot( grid[0,:] ) # plot that uses the first row and 3 cols
ax2 = fig.add_subplot( grid[1:,0:2] )
ax3 = fig.add_subplot( grid[1,2] )
ax4 = fig.add_subplot( grid[2,2] )

ax1.set_title("Plot 1")
ax1.plot([1,3,2,3])

ax2.set_title("Bar Plot")
ax2.bar([1,2,3],[4,2,1])

ax3.set_title("Scatter Plot")
ax3.scatter([1,2,3], [1,2,3])

ax4.set_title("Pie Chart")
ax4.pie([8,7,1]) # sums 8 7 1 and takes percentages
fig.tight_layout()
```

## Types of plots

```python
# linear: plt.plot()
x = np.arange(-10,10,1)
y = x**2
plt.plot(x,y)
# fmt: [color][line style][marker style]
```

```python
# linear: plt.plot()
x = np.arange(-10,10,1)
y = x**2
plt.plot(x,y, "m--D")
# 3rd param: fmt: [color][line style][marker style]
```

```python
x = np.arange(-10,10,1)
y = x**2
plt.plot(x,y, color='#1111FF', linestyle='-', linewidth=4, marker='o', markersize=12,
         markerfacecolor='red', markeredgecolor='black', markeredgewidth=3)
plt.grid(True)
```

```python
# custom grid
x = np.arange(-10,10,1)
y = x**2
plt.plot(x,y, color='#1111FF', linestyle='-', linewidth=4, marker='o', markersize=12,
         markerfacecolor='red', markeredgecolor='black', markeredgewidth=3)
plt.grid(True)
plt.xticks(range(3))
```

```python
# custom grid
x = np.arange(-10,10,1)
y = x**2
plt.plot(x,y, color='#1111FF', linestyle='-', linewidth=4, marker='o', markersize=12,
         markerfacecolor='red', markeredgecolor='black', markeredgewidth=3)
plt.grid(True)
plt.xticks(x)
plt.yticks(y);
```

```python
# custom grid
x = np.arange(-10,10,1)
y = x**2
plt.plot(x,y, color='#1111FF', linestyle='-', linewidth=4, marker='o', markersize=12,
         markerfacecolor='red', markeredgecolor='black', markeredgewidth=3)
plt.grid(True)
plt.xticks(x)
plt.yticks(np.arange(0,110,10));
```
