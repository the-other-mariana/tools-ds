# Manejo de datos
- Extrae las categorías del archivo.


```python
# categorías
import csv 
path='atp_matches_2017.csv'
f = open(path)


years = [2,3,4,5,6,7]
db = []
for i in years:
    f = open(f"atp_matches_201{i}.csv")

    data = csv.reader(f)

    cats = next(data)
    db += list(data)
categories = []
for c in range(len(cats)):
    categories.append(cats[c])
    print(f"index: {c} category: {categories[c]}")
```

- Crea una lista con cada partido, donde los datos de cada partido sean un diccionario. Con las categorías como llave y el valor como el contenido.


```python
# partidos
# each line is a match

match_list = []
print(db[0])
for match in db:
    md = dict()
    for c, m in zip(categories, match):
        md[c] = m
    match_list.append(md)
match_list
```

The zip() function will only iterate over the smallest list passed. If given lists of different lengths, the resulting combination will only be as long as the smallest list passed.

# Filtrado de datos
- Haz un conjunto de todos los nombres de ganadores de partidos.


```python
# ganadores
winners = set()
aux_list = [d['winner_name'] for d in match_list] # or { for ...}
winners = set(aux_list)
winners
```

- Haz un conjunto de todos los nombres de perdedores de partidos.


```python
# perdedores
losers = set()
aux_list = [d['loser_name'] for d in match_list]
losers = set(aux_list)
losers
```

- Enlista qué jugadores no han perdido ningún partido. 

```python
# no han perdido

not_lost = [p for p in list(winners) if p not in losers]
print(not_lost)
```

- ¿Cuántos jugadores no han perdido ningún partido?


```python
# número de jugadores que no han perdido
print(f"{len(not_lost)} players have not lost any game.")
```

- Enlista qué jugadores no han ganado ningún partido.

```python
# no han ganado
not_won = [p for p in list(losers) if p not in winners]
not_won
```

- ¿Cuántos jugadores no han ganado ningún partido?

```python
# número de jugadores que no han ganado
print(f"{len(not_won)} players have not won any game.")
```

- Haz una lista diciendo cuántos partidos ha ganado cada jugador.

```python
# cantidad de partidos ganados por jugador
matches_won = dict()
for w in list(winners):
    matches_won[w] = 0

for match in match_list:
    matches_won[match['winner_name']] += 1
matches_won
```

- Haz una lista diciendo cuántos partidos ha perdido cada jugador

```python
# cantidad de partidos perdidos por jugador

matches_lost = dict()
for l in list(losers):
    matches_lost[l] = 0

for match in match_list:
    matches_lost[match['loser_name']] += 1
matches_lost
```

- Obtén el jugador con más partidos ganados.

```python
# jugador con más partidos ganados
top_winner = tuple()
t = 0
for key in matches_won.keys():
    if t == 0:
        top_winner = tuple([key, matches_won[key]])
    if matches_won[key] > top_winner[1]:
        top_winner = tuple([key, matches_won[key]])
    t += 1
    
print(f"The player with most won matches is: {top_winner[0]} with {top_winner[1]} matches won.")
```

- Obtén el jugador con más partidos perdidos

```python
# jugador con más partidos perdidos
top_loser = tuple()
t = 0
for key in matches_lost.keys():
    if t == 0:
        top_loser = tuple([key, matches_lost[key]])
    if matches_lost[key] > top_loser[1]:
        top_loser = tuple([key, matches_lost[key]])
    t += 1
    
print(f"The player with most lost matches is: {top_loser[0]} with {top_loser[1]} matches lost.")
```

- Obtén todos los jugadores que hayan llegado a semifinales

```python
# Semifinalistas
sf_matches = [f"{m.get('winner_name')}-{m.get('loser_name')}" for m in match_list if m.get('round') == 'SF']
sf_list = set()
for match in sf_matches:
    names = match.split('-')
    sf_list.add(names[0])
    sf_list.add(names[1])
sf_list = list(sf_list)
sf_list
```

- Obtén todos los jugadores que hayan ganado una final.

```python
# Campeones
champions = [p.get('winner_name') for p in match_list if p.get('round') == 'F']
champions
```

# Ordenamiento
- Obtén los 10 jugadores con más partidos ganados y especifica cuántos partidos ganaron.

```python
# top 10
from dataclasses import dataclass
@dataclass 
class Player:
    name:str
    won:int
        
players_won = [Player(key, matches_won[key]) for key in matches_won.keys() if key != '']
sorted_by_won = sorted(players_won, key= lambda x: x.won, reverse=True)
sorted_by_won[:10]
for player in sorted_by_won[:10]:
    print(f"{player.name} won {player.won} games")
```

- Obtén los 10 jugadores con más partidos perdidos y especifica cuántos partidos perdieron.

```python
# top 10
from dataclasses import dataclass
@dataclass 
class Player:
    name:str
    lost:int
        
players_lost = [Player(key, matches_lost[key]) for key in matches_lost.keys() if key != '']
sorted_by_lost = sorted(players_lost, key= lambda x: x.lost, reverse=True)
for player in sorted_by_lost[:10]:
    print(f"{player.name} lost {player.lost} games")
```

- Obtén una lista con los tipos de canchas que existen (pasto, arcilla, etc)

```python
# Tipos de canchas
court_types = {m.get('surface') for m in match_list if m.get('surface') != ''}
court_types = list(court_types)
court_types
```

- Para cada tipo de cancha: Obtén una lista con los tres jugadores con más finales ganadas, y la cantidad de finales ganadas.

```python
from collections import Counter
from dataclasses import dataclass
@dataclass 
class Player:
    name:str
    won:int

winners_surf = []
for c in court_types:
    by_surf = [p.get('winner_name') for p in match_list if p.get('round') == 'F' and p.get('surface') == c]
    winners_surf.append(by_surf)

for i in range(len(winners_surf)):
    d = dict(Counter(winners_surf[i]))
    players_surf = [Player(key, d.get(key)) for key in d.keys() ]
    sort_players = sorted(players_surf, key= lambda x: x.won, reverse=True)
    print(court_types[i])
    for pc in sort_players[:3]:
        print(f"{pc.name} won {pc.won} finals")
    print()
```

# Pregunta ¿Quién fue el mejor jugador de la ATP de 2012 a 2017?

```python
from collections import Counter
from dataclasses import dataclass
@dataclass 
class Player:
    name:str
    won:int

finals_champions = [p.get('winner_name') for p in match_list if p.get('round') == 'F']
d = dict(Counter(finals_champions))
finals_champions = [Player(key, d.get(key)) for key in d.keys() ]
sort_champions = sorted(finals_champions, key= lambda x: x.won, reverse=True)

print(f"The best ATP player from 2012 to 2017 was: {sort_champions[0].name}, with {sort_champions[0].won} finals won")
```
