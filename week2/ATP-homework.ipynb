# Manejo de datos
- Extrae las categorías del archivo.


```python
# categorías
import csv 
path='atp_matches_2017.csv'
f = open(path)
data = csv.reader(f)

cats = next(data)
categories = []
for c in range(len(cats)):
    categories.append(cats[c])
    print(f"index: {c} category: {categories[c]}")
```

- Crea una lista con cada partido, donde los datos de cada partido sean un diccionario. Con las categorías como llave y el valor como el contenido.


```python
# partidos
# each line is a match
import csv 
path='atp_matches_2017.csv'
f = open(path)
data = csv.reader(f)

cats = next(data)
categories = [ c for c in cats ]

match_list = []
for match in data:
    md = dict()
    for i in range(len(categories)):
        md[categories[i]] = match[i]
    match_list.append(md)
match_list
```

# Filtrado de datos
- Haz un conjunto de todos los nombres de ganadores de partidos.


```python
# ganadores
winners = set()
aux_list = [d['winner_name'] for d in match_list]
winners = set(aux_list)
winners
```

- Haz un conjunto de todos los nombres de perdedores de partidos.


```python
# perdedores
losers = set()
aux_list = [d['loser_name'] for d in match_list]
losers = set(aux_list)
losers
```

- Enlista qué jugadores no han perdido ningún partido. 

```python
# no han perdido
not_lost = [p for p in list(winners) if p not in losers]
not_lost
```

- ¿Cuántos jugadores no han perdido ningún partido?


```python
# número de jugadores que no han perdido
print(f"{len(not_lost)} players have not lost any game.")
```

- Enlista qué jugadores no han ganado ningún partido.

```python
# no han ganado
not_won = [p for p in list(losers) if p not in winners]
not_won
```

- ¿Cuántos jugadores no han ganado ningún partido?

```python
# número de jugadores que no han ganado
print(f"{len(not_won)} players have not won any game.")
```

- Haz una lista diciendo cuántos partidos ha ganado cada jugador.

```python
# cantidad de partidos ganados por jugador
matches_won = dict()
for w in list(winners):
    matches_won[w] = 0

for match in match_list:
    matches_won[match['winner_name']] += 1
matches_won
```

- Haz una lista diciendo cuántos partidos ha perdido cada jugador

```python
# cantidad de partidos perdidos por jugador

matches_lost = dict()
for l in list(losers):
    matches_lost[l] = 0

for match in match_list:
    matches_lost[match['loser_name']] += 1
matches_lost
```

- Obtén el jugador con más partidos ganados.

```python
# jugador con más partidos ganados
top_winner = tuple()
t = 0
for key in matches_won.keys():
    if t == 0:
        top_winner = tuple([key, matches_won[key]])
    if matches_won[key] > top_winner[1]:
        top_winner = tuple([key, matches_won[key]])
    t += 1
    
print(f"The player with most won matches is: {top_winner[0]} with {top_winner[1]} matches won.")
```

- Obtén el jugador con más partidos perdidos

```python
# jugador con más partidos perdidos
top_loser = tuple()
t = 0
for key in matches_lost.keys():
    if t == 0:
        top_loser = tuple([key, matches_lost[key]])
    if matches_lost[key] > top_loser[1]:
        top_loser = tuple([key, matches_lost[key]])
    t += 1
    
print(f"The player with most lost matches is: {top_loser[0]} with {top_loser[1]} matches lost.")
```

- Obtén todos los jugadores que hayan llegado a semifinales

```python
# Semifinalistas
sf_matches = [f"{m['winner_name']}-{m['loser_name']}" for m in match_list if m['round'] == 'SF']
sf_list = set()
for match in sf_matches:
    names = match.split('-')
    sf_list.add(names[0])
    sf_list.add(names[1])
sf_list = list(sf_list)
print(sf_list)
```

- Obtén todos los jugadores que hayan ganado una final.

```python
# Campeones
champions = [p['winner_name'] for p in match_list if p['round'] == 'F']
print(champions)
```
