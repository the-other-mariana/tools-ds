```python
import numpy as np

# instead of telling step, you tell number of points
x = np.linspace(0, 5, 100)
y = np.e**x
plt.plot(x,y)
plt.grid(True)

plt.show()
```

```python
import numpy as np

x = np.linspace(0, 5, 100)
y = np.e**x
plt.plot(x,y)
plt.grid(True)

# plot axes (axis)
lims = [0, 4, 0, 140]
plt.axis(lims)
plt.show()

plt.plot(x,y) # you need to put this again to be able to show again
plt.axis([0,4,-5, 15])
plt.show()
```

```python
plt.plot(x,y)
# axis accepts changing xmin,xmax,ymin,ymax
plt.axis(ymin=0)
plt.show()
```

```python
days = np.arange(1,8) # last elem is not included, step default is one
bugs = [10, 8, 3, 4, 11, 2, 3] # a list is not able to multiplt as numpy array
bugs = np.array([10, 8, 3, 4, 11, 2, 3])
plt.plot(days, bugs)
plt.xlabel("Days")
plt.ylabel("Bugs")
plt.title("Axolotl Nutrition")
```

```python
days = np.arange(1,8) 
bugs_ax1 = np.array([10, 8, 3, 4, 11, 2, 3])
bugs_ax2 = np.array([5, 5, 6, 4, 6, 5, 5])
plt.plot(days, bugs_ax1, label="Pancho")
plt.plot(days, bugs_ax2, label="Chencho")
plt.grid(True)
plt.legend()
plt.xlabel("Days")
plt.ylabel("Bugs")
plt.title("Axolotl Nutrition")
```
