## Plot Types
### 1. Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = [1,3,4,5,6,7,9]
y1 = [4,7,2,4,7,8,3]

x2 = [2,4,6,8,10]
y2 = [5,6,2,6,2]

plt.bar(x1,y1, label="Bars No.1")
plt.bar(x2,y2, label="Bars No.2")

plt.legend()
plt.show()
```

```python
# avoid plots blocking another
x1 = [1,3,4,5,6,7,9]
y1 = [4,7,2,4,7,8,3]

x2 = [2,4,6,8,10]
y2 = [5,6,2,6,2]

plt.bar(x1,y1, label="Bars No.1", color=(0,0,1,0.5)) # rgba for opacity
plt.bar(x2,y2, label="Bars No.2", color=(1,0,0,0.5))

plt.xticks(np.arange(1,11))
plt.legend()
plt.show()
```

```python
# label the bars with an specific number
xLabels = "ABCDEFGHIJ"

x1 = [1,3,4,5,6,7,9]
y1 = [4,7,2,4,7,8,3]

x2 = [2,4,6,8,10]
y2 = [5,6,2,6,2]

plt.bar(x1,y1, label="Bars No.1", color=(0,0,1,0.5)) # rgba for opacity
plt.bar(x2,y2, label="Bars No.2", color=(1,0,0,0.5))

plt.xticks(np.arange(1,11), labels = xLabels)
plt.legend()
plt.show()
```

```python
from random import random
def randomColor():
    return (random(), random(), random(), 0.5)

xLabels = "ABCDEFGHIJ"

# default bar width = 1 and sep = 0.5 per side
x1 = np.array([1,3,4,5,6,7,9])
y1 = np.array([4,7,2,4,7,8,3])
z1 = np.array([1,1,1,1,1,1,1]) # decimals can be floats and represent other data
colors = [randomColor() for i in x1]

x2 = np.array([2,4,6,8,10])
y2 = np.array([5,6,2,6,2])

# width depending on other var
#plt.bar(x1,y1,z1, label="Bars No.1", color=colors) # rgba for opacity
# width depending on height
plt.bar(x1,y1,y1/5, label="Bars No.1", color=colors) # rgba for opacity

plt.xticks(np.arange(1,11), labels = xLabels)
plt.legend()
plt.show()
```

```python
from matplotlib import cm # color map
cmap = cm.get_cmap("plasma")

# use a color map
x1 = np.arange(1,20)
y1 = np.arange(1,20)

# normalize 0 to 1
color_n = y1 / max(y1)
color_n = [cmap(c) for c in color_n]

print(cmap(0.5))
plt.bar(x1,y1, color = color_n)
plt.show()
```

```python
x = np.arange(1,20)
y1 = np.arange(1,20)
y2 = np.arange(19,0, -1)

plt.bar(x,y1)
plt.bar(x,y2)
plt.show()
```

```python
x = np.arange(1,20)
y1 = np.arange(1,20)
y2 = np.arange(19,0, -1)

plt.bar(x,y1)
plt.bar(x,y2, bottom=y1) # bottom is a vertical offset

plt.show()
```

```python
x = np.arange(1,20)
y1 = np.arange(1,20)
y2 = np.arange(19,0, -1)

plt.bar(x,y1)
plt.bar(x,y2, bottom=10) # bottom is a vertical offset

plt.show()
```

```python
x = np.arange(1,20)
y1 = np.arange(1,20)
y2 = np.arange(19,0, -1)
y3 = np.random.randint(1,10,19)

plt.bar(x,y1)
plt.bar(x,y2, bottom=y1) 
plt.bar(x,y3, bottom=y1+y2) 

plt.show()
```

```python
# horizontal bars: plt.barh

x = np.arange(1,20)
y1 = np.arange(1,20)

y2 = np.arange(19,0, -1)
y3 = np.random.randint(1,10,19)

plt.barh(x,y1)
plt.barh(x,y2, left=y1) 
plt.barh(x,y3, left=y1+y2) 

plt.show()
```

### 2. Scatter

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(100) # randn is normal distrib, rand is uniform distrib
y = np.random.randn(100)
y2 = np.random.randn(100)
z = np.random.randn(100)
w = np.random.randn(100) * 100 + 10

# plt.plasma()
plt.scatter(x, y, c=z, s=w, marker='^', cmap='plasma') # c takes the array and normalizes it to get a color from a cmap (always normalized)
plt.scatter(x, y2, c=z, s=w, marker='p', cmap='viridis')
plt.show()
```

### 3. Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

x = [8,4,3,1]
plt.pie(x)
plt.show()
```

```python
x = [8,4,3,1]
labels = ['cherry', 'apple', 'avocado', 'oranges']
plt.pie(x, labels = labels)
plt.show()
```

```python
x = np.array([8,4,3,1])
labels = ['cherry', 'apple', 'avocado', 'oranges']

color_n = x / max(x)

cmap = cm.get_cmap("viridis")
color_n = [cmap(c) for c in color_n]
print(color_n)

explode = [0,0,0,0.2] # percentage of explotion
plt.pie(x, labels = labels, explode = explode, shadow=True, autopct='%1.2f%%', colors=color_n) 
# autopct percentage inside format
plt.show()
```

```python
x = np.arange(0,20)
y1 = np.random.randint(5,15,20)
y2 = np.random.randint(3,10,20)
y3 = np.random.randint(2,8,20)

plt.stackplot(x,y1,y2,y3)
plt.show()
```
